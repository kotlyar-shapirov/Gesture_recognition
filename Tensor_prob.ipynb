```python
from torch.utils.data import Dataset, DataLoader

class Dataset_LSTM_Prob(Dataset):
    def __init__(self, data, labels, window_size=30, start_prob=25, stop_prob=85):

        self.data = data
        self.labels = labels
        self.uniq_labels = len(np.unique(self.labels))
        self.ws = window_size
        self.start_prob = start_prob
        self.stop_prob = stop_prob
        
    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        gesture_seqence, label = self.data[idx], self.labels[idx]
        
        prob_sequence = generate_bell(N_tags=self.uniq_labels,
                                      N_time_frames=self.data.shape[1],
                                      cur_tag=label, 
                                      start=self.start_prob, stop=self.stop_prob)
        
#         print(prob_sequence.shape)
#         print(gesture_seqence.shape)
        
        # slicing
        if self.ws < self.data.shape[1]: # window smaller than data length
            start_idx = np.random.randint(low=0, high=gesture_seqence.shape[0]-self.ws)
            gesture_seqence = gesture_seqence[start_idx: start_idx+self.ws]
            prob_sequence = prob_sequence[::,start_idx: start_idx+self.ws]

        gesture_seqence = torch.tensor(gesture_seqence, dtype=torch.float)
        prob_sequence = torch.tensor(prob_sequence, dtype=torch.float).transpose(-1,-2)
        
        return gesture_seqence, prob_sequence
```

### Attention

```python
import torch
import torch.nn as nn
import torch.nn.functional as F


class FeedForward(nn.Module):
    '''simple linear layer + non-linearity'''
    def __init__(self, n_embd_in, n_embd_out, dropout):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(n_embd_in, 4*n_embd_in), # expanding inner layer of information
            nn.ReLU(),
            nn.Linear(4*n_embd_in, n_embd_out), # projection layer as in Multihead attention
            nn.Dropout(dropout)
        )
    def forward(self,x):
        return self.net(x)
    
    
class Block(nn.Module):
    '''attention + linear layer block'''
    def __init__(self, n_embd, n_heads, dropout):
        super().__init__()
        assert n_embd % n_heads == 0, 'the embedding size should be divisible by num heads'
        
        self.Wq = nn.Linear(n_embd, n_embd)
        self.Wk = nn.Linear(n_embd, n_embd)
        self.Wv = nn.Linear(n_embd, n_embd)
        
        self.head = nn.MultiheadAttention(n_embd, num_heads=n_heads,
                                          dropout=dropout, batch_first=True)
        
        self.ffwd = FeedForward(n_embd, n_embd, dropout)
        # layer norm is applied before skipconnection 
        # not as in Attention is All You Need paper 
        self.layernorm1 = nn.LayerNorm(n_embd)
        self.layernorm2 = nn.LayerNorm(n_embd)
        
    def forward(self, x):
        x = self.layernorm1(x)
        
        q, k, v = self.Wq(x), self.Wk(x), self.Wv(x)
        attn_output, attn_output_weights = self.head(q,k,v)
        
        x = x + attn_output
        x = x + self.ffwd(self.layernorm2(x))
        return x

    

class GesTransformer(nn.Module):

    def __init__(self, n_input,
                 n_embd, dropout, n_heads, n_layers, # attention,
                 target_size): 
        super(GesTransformer, self).__init__()
        
        # dropout
        self.dropout_joints= nn.Dropout(0.2)
        # init dim increasing
        self.init_fwd = FeedForward(n_input, n_embd, dropout=0.2)
        
        # positional embedding
        self.position_embedding_table = nn.Embedding(120 + 20, n_embd)
        
        self.blocks = nn.Sequential(*[Block(n_embd, n_heads, dropout) for _ in range(n_layers)])
        self.layer_norm_final = nn.LayerNorm(n_embd)
        self.hidden2tag = nn.Linear(n_embd, target_size)
        
    def forward(self, x, framerate):
        
        B,L,C = x.shape
        # dropout joints
        x = self.dropout_joints(x) 
        
        # initial linear
        x = self.init_fwd(x)
        
        # positional embedding with framerate 
        ticks = torch.arange(L*int(framerate), device=device)[::framerate]
        try:
            pos_x = self.position_embedding_table(ticks)
        except:
            print('ticks shape', ticks.shape)
            print('ticks', ticks)
            print('x shape', x.shape)
            
        x = x + pos_x
        
        # multilayer multihead attention
        x = self.blocks(x) 
        x = self.layer_norm_final(x)
        tags = self.hidden2tag(x)
        
        return tags
        

LSTM_model = GesTransformer(n_input=134,
                            n_embd=256, dropout=0.2, n_heads=2, n_layers=2,
                            target_size=53).to(device)

ttt = torch.rand(16,20,134).to(device)

out = LSTM_model(ttt, 2)
print(out.shape)
```

```python
from torchmetrics import Accuracy, AUROC, F1Score, Precision, Recall, AveragePrecision
from collections import Counter
from sklearn.metrics import accuracy_score

accuracy = Accuracy(task="multiclass", num_classes=54).to(device)
f1_scorer = F1Score(task="multiclass", num_classes=54).to(device)
precision = Precision(task="multiclass", num_classes=54).to(device)
recall = Recall(task="multiclass", num_classes=54).to(device)


def IOU(pred_soft, true_soft, thresh=0.75):
    pred_mask = (pred_soft > thresh).type(torch.int)
    pred_mask[pred_mask == 0] = -1
    true_mask = (true_soft > thresh).type(torch.int)
    i = torch.sum(pred_mask == true_mask)
    u = torch.sum(pred_mask == 1) + torch.sum(true_mask == 1)
    if u != 0:
        return 2*i/u
    else:
        return 1
    
IOU1 = lambda x,y: IOU(x,y, thresh=0.5)
IOU2 = lambda x,y: IOU(x,y, thresh=0.8)

def validate(model, dl_test, metric, framerate):
    res = []
    with torch.no_grad():
        model.eval()
        for seq, tags in dl_test:
            seq, tags = seq.to(device), tags.to(device)
            pred_tags = model(seq, framerate) # added frame coef
#             pred_tags = model(seq)
            
            
            res.append(metric(pred_tags.view(-1,53), tags.view(-1,53)).item())
    model.train(True)
    return np.mean(res)
```
